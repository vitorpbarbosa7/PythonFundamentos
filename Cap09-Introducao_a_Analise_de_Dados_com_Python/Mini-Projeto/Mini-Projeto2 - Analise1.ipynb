## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

fig = plt.figure(figsize = (7,5), dpi = 100)
sns.distplot(df.yearOfRegistration, color = 'g')
# sns.kdeplot(df.yearOfRegistration, shade=True, legend = True, color = 'g')
plt.xlabel('Ano de Registro')
plt.ylabel('Densidade KDE')
plt.title('Densidade de ano de registro dos carros')
```

```python
fig, ax = plt.subplots(figsize = (8,6))
sns.distplot(df.yearOfRegistration, color = '#33cc33', kde = True, ax=ax)
ax.set_title('Distribuição de veículos com base no Ano de Registro')
plt.ylabel('Densidade (KDE)', fontsize = 15)
plt.xlabel('Ano de Registro', fontsize = 15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers

figura2 = plt.figure(figsize = (9,5), dpi = 100)
# sns.boxplot(sns.color_palette('bright'))
sns.boxplot(df.vehicleType, df.price, palette='colorblind')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Intervalo de Preços')
plt.title('Análise de Outliers de preço por tipo de veículo')
```

```python
fig, ax = plt.subplots(figsize = (10,8))
sns.boxplot(df.vehicleType, df.price, palette = 'colorblind')
ax.text(6,28000,'Análise de Outliers',fontsize = 20, color = 'k', ha='center', va = 'center')
ax.xaxis.set_label_text('Tipo de Veículo', fontdict = {'size':14})
ax.yaxis.set_label_text('Range de Preço', fontdict = {'size':14})
plt.show()
```

```python
# Salvando o plot
figura2.savefig("plots/Analise1/price-vehicleType-boxplot.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
df.vehicleType.value_counts()
```

```python
df.vehicleType.value_counts().index
```

```python
figura3 = plt.figure(figsize = (7,5), dpi = 200)
sns.barplot(df.vehicleType.value_counts().index, df.vehicleType.value_counts(), palette ='GnBu_d')
plt.xlabel('Tipos de veículos')
plt.xticks(fontsize = 8)
plt.ylabel('Número de veículos')
plt.yticks(fontsize = 8)
plt.title('Número de veículos à venda por tipo')
```

```python
g = sns.factorplot(x = 'vehicleType', data = df, kind = 'count', palette = 'GnBu_d', size = 6, aspect = 1.5)
g.ax.xaxis.set_label_text('Tipo de veículo', fontdict = {'size':16})
g.ax.yaxis.set_label_text('Número de veículos', fontdict= {'size':16})
g.ax.set_title('Contagem total de veículos à venda por tipo de veículo')
#Para poder colocar label em todas barras:
for barra in g.ax.patches:
    g.ax.annotate((barra.get_height()), (barra.get_x()+0.1, barra.get_height()+500))
```

```python
# Salvando o plot
figura3.savefig("plots/Analise1/count-vehicleType.png")
```
