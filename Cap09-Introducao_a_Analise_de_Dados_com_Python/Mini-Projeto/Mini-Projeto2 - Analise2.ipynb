```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Número de veículos pertencentes a cada marca

```python
list(df)
```

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
figura1 = plt.figure(figsize = (5,7), dpi = 150)
sns.barplot(df.brand.value_counts(), df.brand.value_counts().index,palette = 'Reds_d')
plt.yticks(fontsize = 6)
plt.xticks(fontsize = 6)
```

```python
g = sns.factorplot(y = 'brand', data = df, kind = 'count', palette= 'Reds_r', size = 8, aspect = 1.7)
g.ax.set_title('Veículos por marca')
g.ax.xaxis.set_label_text('brand')
g.ax.yaxis.set_label_text('Marca')
```

```python
# Salvando o plot
figura1.savefig(("plots/Analise2/brand-vehicleCount.png"))
```
