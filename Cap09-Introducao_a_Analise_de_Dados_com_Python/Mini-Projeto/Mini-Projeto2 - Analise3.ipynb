```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
list(df)
```

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

figure1 = plt.figure(figsize = (10,6))
sns.barplot(x=df['fuelType'], y = df['price'], hue = df['gearbox'], palette ='bright')
plt.xlabel('Tipo de combustível')
plt.ylabel('Preço Médio')
plt.title('Preço Médio de combustível por tipo de câmbio')
plt.show()
```

```python
# Salvando o plot
figure1.savefig("plots/Analise3/fueltype-vehicleType-price2.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

figure2 = plt.figure(figsize = (10,6))
sns.barplot(x=df['vehicleType'], y = df['powerPS'], hue = df['gearbox'], palette ='bright')
plt.xlabel('Tipo de veículo')
plt.ylabel('Preço Médio')
plt.title('Preço Médio de combustível por tipo de veículo')
plt.show()
```

```python
# Salvando o plot
figure2.savefig("plots/Analise3/vehicletype-fueltype-power2.png")
```
